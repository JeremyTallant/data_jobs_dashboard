```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
import re
```

```python
#Connect to local database
protocol = 'postgresql'
username = 'postgres'
password = 'postgres'
host = 'localhost'
port = 5432
database_name = 'jobs'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
df = pd.read_sql('select * from jobs', con=engine)
df
```

```python
# Count the amount of times certain data analysis tools appear in the job descripiton
python_counts = df['Job description'].str.count(r'\bPython\b', flags=re.IGNORECASE).sum()
sql_counts = df['Job description'].str.count(r'\bSQL\b', flags=re.IGNORECASE).sum()
nosql_counts = df['Job description'].str.count(r'\bNoSQL\b', flags=re.IGNORECASE).sum()
excel_counts = df['Job description'].str.count(r'\bExcel\b', flags=re.IGNORECASE).sum()
r_counts = df['Job description'].str.count(r'\bR\b', flags=re.IGNORECASE).sum()
tableau_counts = df['Job description'].str.count(r'\bTableau\b', flags=re.IGNORECASE).sum()
powerbi_counts = df['Job description'].str.count(r'\bPower BI\b|\bPowerBI\b', flags=re.IGNORECASE).sum()
javascript_counts = df['Job description'].str.count(r'\bjavascript\b', flags=re.IGNORECASE).sum()
html_counts = df['Job description'].str.count(r'\bHTML\b', flags=re.IGNORECASE).sum()
css_counts = df['Job description'].str.count(r'\bCSS\b', flags=re.IGNORECASE).sum()
mongodb_counts = df['Job description'].str.count(r'\bMongoDB\b', flags=re.IGNORECASE).sum()
```

```python
# Make a dictionary and then convert to Data Frame
counts_dict = {'Data Analysis Tools': ['Python', 'SQL', 'NoSQL', 'Excel', 'R', 'Tableau', 'PowerBI', 'JavaScript', 'HTML', 'CSS', 'MongoDB'],
               'Count': [python_counts, sql_counts, nosql_counts, excel_counts, r_counts, tableau_counts, powerbi_counts, javascript_counts, html_counts, css_counts, mongodb_counts]}
counts_df = pd.DataFrame(counts_dict)
counts_df
```

```python
sorted_counts_df = counts_df.sort_values('Count', ascending=False)
sorted_counts_df
```

```python
# Load the dataframe into sql database
sorted_counts_df.to_sql(name='tools', con=engine, if_exists='replace', index=False)
```
